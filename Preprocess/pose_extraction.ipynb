```python
import os
import subprocess
import logging

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

video_root = "/content/drive/MyDrive/MajorProject/Video"
data_root = "/content/drive/MyDrive/MajorProject/Data"

openpose = "/content/openpose/build/examples/openpose/openpose.bin"

for folder in os.listdir(video_root):

    video_folder = os.path.join(video_root, folder)

    if not os.path.isdir(video_folder):
        continue

    data_folder = os.path.join(data_root, folder)

    # Skip if folder already exists in Data
    if os.path.exists(data_folder):
        print("Skipping folder (already processed):", folder)
        continue

    pose_folder = os.path.join(data_folder, "pose")
    os.makedirs(pose_folder, exist_ok=True)

    for video in os.listdir(video_folder):

        if not video.endswith(".mp4"):
            continue

        video_path = os.path.join(video_folder, video)
        logging.info("Processing: %s", video_path)

        print("Processing:", video_path)

        cmd = [
            openpose,
            "--video", video_path,
            "--write_json", pose_folder,
            "--display", "0",
            "--render_pose", "0",
            "--model_pose", "BODY_25",
            "--number_people_max", "1",
            "--net_resolution", "-1x256"
        ]

        subprocess.run(cmd)
```
